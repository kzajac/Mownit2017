Do poszukiwania pierwiastków funkcji w Julii używamy pakietu Roots

```julia
using Roots  
using Plots
```

```julia
# przykładowa funkcja do testów
f(x) = cos(x) - x
plot(f, -2, 2)
```

Funkcja <i> fzero</i>  dobiera odpowiedni algorytm w zależności od sposobu wywołania.


Metoda bisekcji - jesli wywołamy fzero z podaniem  przedziału (tutaj [0,1])

```julia
x = fzero(f, 0, 1, verbose=true)  
```

Innym typem metody jest metoda Newtona, potrzebująca punktu startowego i wykorzystująca pochodną funkcji.

```julia
# metoda Newtona funkcja, punkt startowy (tutaj x0=0)
newton(f, 0, verbose=true)
```

Metoda Newtona korzysta z pakietu ForwardDiff, aby obliczyć pochodną funkcji. Pakiet ten udostępnia dla uzytkownika   operator D, który zwraca funkcję pochodną

```julia
# przykład użycia wprost :
# obliczamy wartości x^2 dla liczb 1..10
x=linspace(1,10,10)
y=(x->x^2).(x)
```

```julia
# obliczamy wartości funkcji pochodnej dla x^2 czyli 2x:
yprim=D(x->x^2).(x)
```

```julia
# można tez samemu podać pochodną
newton(f, D(f),0, verbose=true)
```

Metoda siecznych

```julia
# wywołanie fzero używając punktu startowego (a nie przedziału) oraz opcji order=1   wykorzystuje metodę siecznych  - #podajemy punkt startowy tutaj x0=0 oraz order=1
x = fzero(f, 0,order=1, verbose=true) 
```

```julia
# metoda siecznych może być wywołana bezpośrednio - wtedy podajemy przedział
secant_method(f,0,1, verbose=true)
```

Metody fzero dla wyzszych rzędów to wariacje metody Newtona, które nie korzystaja z pochodnej, ale ja przyblizają.


```julia
# np. metoda Steffensena przybliża pochodną poprzez  (f(x + f(x)) - f(x))/f(x) 
# aby ja wywołać podajemy punkt startowy oraz order 2
x = fzero(f, 0, order=2, verbose=true) 
```

```julia
# mozliwe sa jeszcze rzędy 5, 8 i 16
x = fzero(f, 0, order=16, verbose=true) 
```

Metoda Halleya używa pierwszej i drugiej pochodnej

```julia
halley(f, 2, verbose=true)
```

Można użyć funkcji fzero do znajdowania nieciagłości

```julia
plot(x -> 1/x)
```

```julia
fzero(x -> 1/x, -1, 1, verbose=true)
```

fzeros - szukanie więcej niż jednego pierwiastka. Wykorzystuje podział przedziału na mniejsze podprzedziały

```julia
fzeros(x ->(x-3)*x, -10, 10,verbose=true)
```

```julia
plot(x ->(x-3)*x,-1,4)
```
